```python
#selecting features

from sklearn.feature_selection import SelectKBest, chi2, f_classif, mutual_info_classif

#chi2 is chi square test
#f_classif is anova F 
#mutual info.. self explanatory 

selector = SelectKBest(score_func = f_classif, k = 10)
X_new = selector.fit_transform(X,y)

selected_indices = selector.get_support(indices=True)
selected_features = X.columns[selected_indices]

print("Selected Features:")
print(selected_features)
```

scaling data, min max scaling technique

X_normalized = (X - X_min) / (X_max - X_min)

```python
#malign
xm_axis = df_v1.query("Diagnosis=='M'")
xm_axis = xm_axis['radius1']
xm_axis
```

```python
#benign
xb_axis = df_v1.query("Diagnosis=='B'")
xb_axis = xb_axis['radius1']
xb_axis
```

```python
#scaled 0-1 data
min_x = np.min(xb_axis)

max_x = np.max(xb_axis)

scaled_data = (xb_axis-min_x) / (max_x - min_x)

sns.displot(scaled_data)
```

```python
#standardized between 0 and 1
min_x = np.min(xm_axis)

max_x = np.max(xm_axis)

scaled_data = (xm_axis-min_x) / (max_x - min_x)

sns.displot(scaled_data)
```

x_z-score_normalized = (x-x_mean)/x_std

```python
#z score normalized graph
mean_x = np.mean(xm_axis)
std_dev_x = np.std(xm_axis)

standardized_x = (xm_axis-mean_x) / std_dev_x

sns.displot(standardized_x)

counts = len(xm_axis)
```

```python
#normalized graph
mean_x = np.mean(xb_axis)
std_dev_x = np.std(xb_axis)

standardized_x = (xb_axis-mean_x) / std_dev_x

sns.displot(standardized_x)
```
